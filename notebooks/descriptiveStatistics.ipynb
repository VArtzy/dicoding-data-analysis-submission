```python
import numpy as np

jumlah_kucing = np.array([3, 2, 1, 1, 2, 3, 2, 1, 0, 2])
jumlah_kucing.mean()
```

```python
np.median(jumlah_kucing)
```

```python
from scipy import stats
stats.mode(jumlah_kucing)[0]
```

```python
jumlah_kucing.max() - jumlah_kucing.min()
```

```python
np.percentile(jumlah_kucing, 75) - np.percentile(jumlah_kucing, 25)
```

```python
import pandas as pd

jumlah_kucing_series = pd.Series(jumlah_kucing)
jumlah_kucing_series.var()
```

```python
jumlah_kucing_series.std()
```

```python
import matplotlib.pyplot as plt

plt.hist(jumlah_kucing, bins=4)
plt.show()
```

```python
jumlah_kucing_series.skew()
```
