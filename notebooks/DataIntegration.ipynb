```python
import kagglehub
import pandas as pd

path = kagglehub.dataset_download("wowevan/dataset-kesejahteraan-pekerja-indonesia")

gk = pd.read_csv(path + "/garisKemiskinan.csv")
peng = pd.read_csv(path + "/pengeluaran.csv")
ump = pd.read_csv(path + "/minUpah.csv")
avg = pd.read_csv(path + "/rataRataUpah.csv")
```

```python
#Data Selection
gk = gk.filter(items=['provinsi', 'tahun', 'gk', 'periode'])
ump = ump.filter(items=['provinsi', 'tahun', 'ump'])
peng = peng.filter(items=['provinsi', 'tahun', 'peng'])
avg = avg.filter(items=['provinsi', 'tahun', 'upah'])
```

```python
#Data Integration
merged_data = gk.merge(ump, on=['provinsi', 'tahun']).merge(peng, on=['provinsi', 'tahun']).merge(avg, on=['provinsi', 'tahun'])
```
